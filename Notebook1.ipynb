# Quality control

```bash

total_reads=$(samtools view -c S1/S1_sorted.bam)

on_target_reads=$(bedtools intersect -a S1/S1_sorted.bam -b AODABCV3R7.designed.bed  -u | samtools view -c -)

percentage_on_target=$(echo "scale=2; ($on_target_reads / $total_reads) * 100" | bc)

echo "Доля целевых выровненных прочтений: $percentage_on_target%"
echo "Общее число прочтений: $total_reads"
```
